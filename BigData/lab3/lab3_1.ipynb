## Docker + Hive

### Запуск контейнеров
```
docker compose up -d
```

### Загрузка CSV в HDFS

```
docker exec -it namenode bash
hdfs dfs -mkdir -p /steam
hdfs dfs -put /host/data/steam_reviews.csv /steam/
exit

```
### Запуск SQL в Hive
```
docker exec -it hive bash
hive -f /host/hive_query.hql
```

```
docker compose down
```
