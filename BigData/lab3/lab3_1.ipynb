## Docker Hadoop + Hive

### Запуск контейнеров
```
docker compose up -d
```

### Загрузка CSV в HDFS
```
# 1. Скопировать файл в контейнер
docker cp data/steam_reviews.csv namenode:/steam_reviews.csv

# 2. Загрузить в HDFS
docker exec -it namenode hdfs dfs -mkdir -p /steam
docker exec -it namenode hdfs dfs -put /steam_reviews.csv /steam/

# 3. Проверить
docker exec -it namenode hdfs dfs -ls /steam

```
### Запуск SQL в Hive
```
docker cp hive_query.hql hive-server:/hive_query.hql

 docker exec -it hive-server /opt/hive/bin/beeline -u jdbc:hive2://localhost:10000 -f /hive_query.hql
```
